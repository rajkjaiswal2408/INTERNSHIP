```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://presidentofindia.nic.in/former-presidents.htm"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

President_Name = []
Office_term = []

for i in soup.find_all('div', class_="presidentListing"):
    President_Name.append(i.text.split('(')[0].replace('\n',''))
for i in soup.find_all('div', class_="presidentListing"):
    Office_term.append(i.text.split(':')[1].replace('\n', '').replace('http',''))

data = {"Name": President_Name, "Term of Office": Office_term}
df = pd.DataFrame(data)
df.index +=1
df
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests


page=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")


soup=BeautifulSoup(page.content,'html.parser')


Team=[]
for i in soup.find_all("span",class_="u-hide-phablet"):
    Team.append(i.text)


Matches=[]
Points=[]
ODI_MMatches=soup.find_all('td',class_="table-body__cell u-center-text")
for i in range(0,len(ODI_MMatches),2):
    Matches.append(ODI_MMatches[i].get_text().replace(',',''))
    Points.append(ODI_MMatches[i+1].get_text().replace(',',''))
match_1 = soup.find("td",class_="rankings-block__banner--matches").get_text()
Matches.insert(0,match_1)
points_1 = soup.find('td',class_="rankings-block__banner--points").get_text().replace(',','')
Points.insert(0,points_1)


Rating=[]
rating_all = soup.find_all('td', class_="table-body__cell u-text-right rating")
for i in rating_all:
    Rating.append(i.get_text().replace('\n',''))
Rating_1 = soup.find('td', class_="rankings-block__banner--rating u-text-right").get_text().replace('\n','').strip()
Rating.insert(0,Rating_1)


ICC_top_10_MTeams = pd.DataFrame({})
ICC_top_10_MTeams['Team'] = Team[0:10]
ICC_top_10_MTeams['Matches'] = Matches[0:10]
ICC_top_10_MTeams['Points'] = Points[0:10]
ICC_top_10_MTeams['Rating'] = Rating[0:10]

ICC_top_10_MTeams.index +=1


ICC_top_10_MTeams
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = "https://www.cnbc.com/world/?region=world"
response = requests.get(url)


soup = BeautifulSoup(response.content, 'html.parser')


headlines = []
times = []
news_links = []

articles = soup.find_all(class_="Card-titleContainer")


for article in articles:
    headline = article.find(class_="Card-title").text.strip()
    headlines.append(headline)
    
    
    time_element = article.find(class_="Card-time")
    if time_element:
        time = time_element.text.strip()
    else:
        time = None
    times.append(time)
    
    
    news_link_element = article.find('a')
    if news_link_element:
        news_link = news_link_element['href']
    else:
        news_link = None
    news_links.append(news_link)

data = {
    'Headline': headlines,
    'Time': times,
    'News Link': news_links}
df = pd.DataFrame(data)
df.index +=1
df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd



page=requests.get("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
page



soup=BeautifulSoup(page.text,'html.parser')
soup




Paper_Title=[]
Authors=[]
Pub_Date=[]
Paper_URL=[]


Papers=soup.find_all('h2',class_='sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg')
for i in Papers:
    Paper_Title.append(i.get_text())
    

author=soup.find_all('span',class_='sc-1w3fpd7-0 dnCnAO')
for i in author:
    Authors.append(i.get_text())
    

PubDate=soup.find_all('span',class_='sc-1thf9ly-2 dvggWt')
for i in PubDate:
    Pub_Date.append(i.get_text())


soup2=soup.find('div',class_='sc-orwwe2-3 jOMrrY').find_all('a')
for i in soup2:
    Paper_URL.append(i.get('href', None))


Most_Downloaded=pd.DataFrame({})
Most_Downloaded['Title']=Paper_Title
Most_Downloaded['Author']=Authors
Most_Downloaded['Publish Date']=Pub_Date
Most_Downloaded['URL']=Paper_URL
Most_Downloaded.index +=1
Most_Downloaded
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd



page=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
page



soup=BeautifulSoup(page.text,'html.parser')
soup




Rest_Name=[]
Cuisine=[]
Location=[]
Rating=[]
Image_URL=[]


Name=soup.find_all('a',class_='restnt-name ellipsis')
for i in Name:
    Rest_Name.append(i.get_text())
    

cui=soup.find_all('span',class_='double-line-ellipsis')
for i in cui:
    Cuisine.append(i.text.split('|')[1])

Loc=soup.find_all('div',class_='restnt-loc ellipsis')
for i in Loc:
    Location.append(i.get_text())
    

Rate=soup.find_all('div',class_='restnt-rating rating-4')
for i in Rate:
    Rating.append(i.get_text())
    
imgurl=soup.find_all('img',class_='no-img')
for i in imgurl:
    Image_URL.append(i.get('data-src'))
    




Rest_Data=pd.DataFrame({})
Rest_Data['Title']=Rest_Name
Rest_Data['Cuisine']=Cuisine
Rest_Data['LOcation']=Location
Rest_Data['Rating']=Rating
Rest_Data['ImageURL']=Image_URL
Rest_Data.index +=1
Rest_Data
```
